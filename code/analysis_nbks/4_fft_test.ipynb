```python
plt.figure()
x=np.arange(0,5*np.pi,np.pi/16)
y=np.sin(x)
plt.plot(x,y,marker='D',linestyle='-.')
```

```python
y=np.sin(x)+3*np.sin(2*x)+5*np.sin(7*x)
plt.figure()
plt.plot(x,y,marker='*', linestyle='-')
```

```python
ft=np.fft.fft(y)
freq=np.fft.fftfreq(y.shape[-1])
plt.figure()
plt.plot(freq,ft.real,freq,ft.imag,marker='o',linestyle='')
```
