```python
# Load dat
df1=pd.DataFrame([])
lst=['tl','ntl_1copy','ntl_2copy','ntl_3copy','ntl_4copy','tl_3lvl','tl_4lvl']

for i in lst: 
    fname='../stored_results/3_64_ntl/results_scaling_{0}.txt'.format(i)
    # a1=np.loadtxt(fname,delimiter='\t')
    cols=['L','m','levels','n_per_level','conv_iters']
    df=pd.read_csv(fname,delimiter='\t',names=cols)
    df=df.sort_values(by=['levels','n_per_level','L','m'])
    df['run_type']=i
    df1=df1.append(df)
df1
```

```python
dfa=df1[['run_type','m','conv_iters']]
dfa
```

```python
labels=np.unique(dfa.run_type.values)
plt.figure()
for i,marker in zip(labels,['*','D','H','x','d','h','p']):
    df=dfa[dfa.run_type==i][['m','conv_iters']]
    # display(df)
    plt.plot(df.m.values,df.conv_iters.values*3,label=i,marker=marker)
plt.xticks(df.m.values)
plt.yscale('log')
plt.xlabel('mass')
plt.ylabel('Iterations for convergence')
plt.legend()
plt.show()
```
